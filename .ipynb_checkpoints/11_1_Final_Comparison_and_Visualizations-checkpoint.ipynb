```python
data_dir = "/Users/samskanderi/Documents/Research_Project/SWANSF/code/results/"
with open(data_dir + 'SVM_RusTomekSmote_WithoutB&C_NewFeatures_Results' + ".pkl", 'rb') as f:
    svm_newf_RTS=np.array(pickle.load(f))
with open(data_dir + 'SVM_RusTomekAdasyn_WithoutB&C_NewFeatures_Results' + ".pkl", 'rb') as f:
    svm_newf_RTA=np.array(pickle.load(f))
with open(data_dir + 'MLPClassifier_RusTomekSmote_WithoutB&C_NewFeatures_Results' + ".pkl", 'rb') as f:
    mlp_newf_RTS=np.array(pickle.load(f))
with open(data_dir + 'MLPClassifier_RusTomekAdasyn_WithoutB&C_NewFeatures_Results' + ".pkl", 'rb') as f:
    mlp_newf_RTA=np.array(pickle.load(f))
with open(data_dir + 'SVM_Smote_WithoutB&C_NewFeatures_Results' + ".pkl", 'rb') as f:
    svm_newf_S=np.array(pickle.load(f))
with open(data_dir + 'SVM_Adasyn_WithoutB&C_NewFeatures_Results' + ".pkl", 'rb') as f:
    svm_newf_A=np.array(pickle.load(f))
with open(data_dir + 'MLPClassifier_Smote_WithoutB&C_NewFeatures_Results' + ".pkl", 'rb') as f:
    mlp_newf_S=np.array(pickle.load(f))
with open(data_dir + 'MLPClassifier_Adasyn_WithoutB&C_NewFeatures_Results' + ".pkl", 'rb') as f:
    mlp_newf_A=np.array(pickle.load(f))
    
#with open(data_dir + 'Rocket_RusTomekSmote_WithoutB&C_NewFeatures_Results' + ".pkl", 'rb') as f:
    #rocket_newf_RTS=np.array(pickle.load(f))
#with open(data_dir + 'Rocket_RusTomekAdasyn_WithoutB&C_NewFeatures_Results' + ".pkl", 'rb') as f:
    #rocket_newf_RTA=np.array(pickle.load(f))
#with open(data_dir + 'TSF_RusTomekSmote_WithoutB&C_NewFeatures_Results' + ".pkl", 'rb') as f:
    #tsf_newf_RTS=np.array(pickle.load(f))
#with open(data_dir + 'TSF_RusTomekAdasyn_WithoutB&C_NewFeatures_Results' + ".pkl", 'rb') as f:
    #tsf_newf_RTA=np.array(pickle.load(f))
    
with open(data_dir + 'SVM_NewFeatures_Results' + ".pkl", 'rb') as f:
    svm_newf=np.array(pickle.load(f))
with open(data_dir + 'MLPClassifier_NewFeatures_Results' + ".pkl", 'rb') as f:
    mlp_newf=np.array(pickle.load(f))
#with open(data_dir + 'Rocket_NewFeatures_Results' + ".pkl", 'rb') as f:
    #rocket_newf=np.array(pickle.load(f))
#with open(data_dir + 'TSF_NewFeatures_Results' + ".pkl", 'rb') as f:
    #tsf_newf=np.array(pickle.load(f))


names = ['SVM', 'SVM (Smote)', 'SVM (Adasyn)', 'SVM (RUS-Tomek-Smote)', 'SVM (RUS-Tomek-Adasyn)',
         'MLP', 'MLP (Smote)', 'MLP (Adasyn)', 'MLP (RUS-Tomek-Smote)', 'MLP (RUS-Tomek-Adasyn)']
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("darkgrid")

values = np.array([svm_newf[:,6], svm_newf_S[:,6], svm_newf_A[:,6], svm_newf_RTS[:,6],
                   svm_newf_RTA[:,6], mlp_newf[:,6], mlp_newf_S[:,6], 
                   mlp_newf_A[:,6], mlp_newf_RTS[:,6],
                   mlp_newf_RTA[:,6]])

import pandas as pd
df = pd.DataFrame(values.reshape(-1, 1), columns=["TSS"])
df["Model"] = np.repeat(names, values.shape[1])
df["TrainTestCombo"] = np.tile(np.arange(values.shape[1]), values.shape[0])


# Create a beautiful violin plot with Seaborn
plt.figure(figsize=(10, 10))
sns.boxplot(data=df, x="TSS", y="Model", width=0.4)

# Customize the plot
plt.title("TSS Evaluation of Different ML Models After and Before Sampling")
plt.xlabel("TSS Value")
plt.ylabel("ML Models")
plt.grid(axis="y")

ax = plt.gca()
ax.set_xlim(-.05, 1.05)  # Set x-axis limits
# Show the plot
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("darkgrid")

values = np.array([svm_newf[:,6], svm_newf_S[:,6], svm_newf_A[:,6], svm_newf_RTS[:,6],
                   svm_newf_RTA[:,6], mlp_newf[:,6], mlp_newf_S[:,6], 
                   mlp_newf_A[:,6], mlp_newf_RTS[:,6],
                   mlp_newf_RTA[:,6]])

import pandas as pd

df = pd.DataFrame(values, columns=[str(i) for i in range(1, values.shape[1] + 1)])

# Create a beautiful violin plot with Seaborn
plt.figure(figsize=(10, 6))
for i in range(0, df.shape[0]):
    if i % 5 == 0:
        sns.lineplot(data=df.iloc[i], dashes=True, label=names[i], linestyle=':', linewidth=1)
    elif i % 5 == 1 or i % 5 == 2:
        sns.lineplot(data=df.iloc[i], dashes=True, label=names[i], linestyle='--', linewidth=1)
    elif i % 5 == 3 or i % 5 == 4:
        sns.lineplot(data=df.iloc[i], dashes=True, label=names[i], linestyle='-', linewidth=1)


# Customize the plot
plt.title("TSS Evaluation of Different ML Models After and Before Sampling")
plt.xlabel("Train-Test Combinations")
plt.ylabel("TSS Value")
plt.grid(axis="y")

# Show the plot
plt.legend(fontsize='x-small')

plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set_style("darkgrid")

values = {
    'Y' : [np.median(svm_newf[:,6]), np.median(svm_newf_S[:,6]), np.median(svm_newf_A[:,6]), np.median(svm_newf_RTS[:,6]),
                   np.median(svm_newf_RTA[:,6]), np.median(mlp_newf[:,6]), np.median(mlp_newf_S[:,6]), 
                   np.median(mlp_newf_A[:,6]), np.median(mlp_newf_RTS[:,6]),
                   np.median(mlp_newf_RTA[:,6])],
    
    'X' : [np.median(svm_newf[:,10]), np.median(svm_newf_S[:,10]), np.median(svm_newf_A[:,10]), np.median(svm_newf_RTS[:,10]),
                   np.median(svm_newf_RTA[:,10]), np.median(mlp_newf[:,10]), np.median(mlp_newf_S[:,10]), 
                   np.median(mlp_newf_A[:,10]), np.median(mlp_newf_RTS[:,10]),
                   np.median(mlp_newf_RTA[:,10])],
    'Class' : ['Without Sampling', 'Over Sampling', 'Over Sampling', 'Both Over & Under Sampling'
              ,'Both Over & Under Sampling', 'Without Sampling', 'Over Sampling', 'Over Sampling'
               , 'Both Over & Under Sampling', 'Both Over & Under Sampling'],
}


df = pd.DataFrame(values)

# Create a beautiful violin plot with Seaborn
plt.figure(figsize=(6, 4))

sns.scatterplot(data=df[df['Class'] == 'Without Sampling'], x='X', y='Y', label='Without Sampling', marker='^')

# Plot Class B in red with squares
sns.scatterplot(data=df[df['Class'] == 'Over Sampling'], x='X', y='Y', label='Over Sampling', marker='o')

# Plot Class C in green with triangles
sns.scatterplot(data=df[df['Class'] == 'Both Over & Under Sampling'], x='X', y='Y', label='Both Over & Under Sampling', marker='s')


# Customize the plot
plt.title("Evaluation of Different ML Models After and Before Sampling (Median Values of 10 train-test cominations)")
plt.xlabel("Recall")
plt.ylabel("TSS")

ax = plt.gca()
ax.set_xlim(0, 1)  # Set the x-axis limits from 2 to 7
ax.set_ylim(0, .85)  # Set the y-axis limits from 3 to 8
plt.grid(axis="y")

# Show the plot
plt.legend(fontsize='x-small')

plt.show()
```
